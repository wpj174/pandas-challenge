```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.count()
```

```python
# Get list of unique players
unique_players = purchase_data.drop_duplicates(subset='SN')

unique_players.count()
```

```python
# Establish bin age ranges and bin names
bins = [0,9,14,19,24,29,34,39,100]
bin_names = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase_age = unique_players[["SN", "Age"]].copy()
purchase_age["Age Group"] = pd.cut(purchase_age["Age"], bins, labels=bin_names, include_lowest=True)
#age_groups = pd.cut(purchase_age["Age"], bins, labels=bin_names, include_lowest=True)
#age_groups
#purchase_age["Age Groups"] = age_groups
#unique_players["Age"]
purchase_age.head()
#purchase_age.count()
#purchase_age.loc[:,"Age"]
```
